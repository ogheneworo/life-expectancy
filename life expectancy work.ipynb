## Training a Linear Regreession Model

```python
life_e.columns
```

## spliting our viarables to X and y 

```python
X=life_e[['Income', 'Schooling','BMI','Polio','Diphtheria ', 'Alcohol','Health expenditure percentage','Hepatitis B','Measles ','infant deaths','under-five deaths ', 'Thinness 10-19 years','Thinness 5-9 years','Adult Mortality','Diphtheria ', ' HIV/AIDS']]
y=life_e['Life expectancy ']
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

## Creating and Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

## Model Evaluation
#### let's evaluate the model by checking out it's coefficient and how we can interpret them.

```python
print (lm.intercept_)
```

```python
lm.coef_
```

```python
X_train.columns
```

```python
a = lm.coef_

a
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns, columns=['Coefficient'])
#create a data frame lm.coef is the data x.columns is the index
coeff_df
```

```python
plt.figure(figsize =(15,6))
sns.barplot(x= X.columns, y=lm.coef_)
plt.xticks(rotation=-45)
```

## Prediction from our Model

```python
lm.score(X_test,y_test)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=50)
```

```python
from sklearn import metrics
```

```python
print('MAE',metrics.mean_absolute_error(y_test,predictions))
print('MSE',metrics.mean_squared_error(y_test,predictions))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
metrics.explained_variance_score(y_test,predictions)
```
